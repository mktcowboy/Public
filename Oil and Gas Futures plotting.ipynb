```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
CL = yf.Ticker("CL=F")
df1 = CL.history(start="2020-01-01")
df1 = df1.drop(['Dividends','Stock Splits'], axis=1)
df1.tail()
```

```python
NG = yf.Ticker("NG=F")
df2 = NG.history(start="2020-01-01")
df2 = df2.drop(['Dividends','Stock Splits'], axis=1)
# tail is most recent data
df2.tail()
```

```python
# PLOTTING BOTH TOGETHER

# Set the figure size
ax = plt.figure(figsize=(15, 7))

# Plot both futures close prices
ax = df1['Close'].plot(label='Oil Futs', color='b')
ax2 = df2['Close'].plot(secondary_y=True, color='g',  ax=ax, label='Gas Futs')

# Set the title and axis labels and sizes
plt.title('Oil andd Gas Futs', fontsize=16)
ax.set_xlabel('Year-Month', fontsize=15)
ax.set_ylabel('Oil Price ($)', fontsize=15)
ax2.set_ylabel('Gas Price ($)', fontsize=15)
ax.tick_params(axis='both', labelsize=15)
ax2.tick_params(axis='y', labelsize=15)
h1, l1 = ax.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax.legend(h1+h2, l1+l2, loc=2, prop={'size': 15})

# Show the plot
```

```python
# PLOTTING BOTH TOGETHER

# Set the figure size
ax = plt.figure(figsize=(15, 7))

# Filter the data to start on July 1, 2020
df1_filtered = df1[df1.index >= '2020-07-01']
df2_filtered = df2[df2.index >= '2020-07-01']

# Plot both futures close prices
ax = df1_filtered['Close'].plot(label='Oil Futs', color='black')
ax2 = df2_filtered['Close'].plot(secondary_y=True, color='green',  ax=ax, label='Gas Futs')

# Set the title and axis labels and sizes
plt.title('Oil & Gas Futs', fontsize=16)
ax.set_xlabel('Year-Month', fontsize=15)
ax.set_ylabel('Oil Price ($)', fontsize=15)
ax2.set_ylabel('Gas Price ($)', fontsize=15)
ax.tick_params(axis='both', labelsize=15)
ax2.tick_params(axis='y', labelsize=15)
h1, l1 = ax.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax.legend(h1+h2, l1+l2, loc=2, prop={'size': 15})

# Show the plot
```

```python
# Calculate the ratio of oil to gas futures
df1['Ratio'] = df1['Close'] / df2['Close']

# Set the figure size
fig, ax = plt.subplots(figsize=(15, 7))

# Plot the ratio
ax.plot(df1.index, df1['Ratio'], color='b', label='Oil/Gas Ratio')

# Set the title and axis labels and sizes
ax.set_title('Oil/Gas Futures Ratio', fontsize=16)
ax.set_xlabel('Year-Month', fontsize=15)
ax.set_ylabel('Ratio', fontsize=15)
ax.tick_params(axis='both', labelsize=15)
ax.legend(loc='upper left', fontsize=15)

# Show the plot
plt.show()
```

```python
# Calculate the ratio of oil to gas futures
df1['Ratio'] = df1['Close'] / df2['Close']

# Filter the data to start from July 2020
df1_filtered = df1[df1.index >= '2020-07-01']
df2_filtered = df2[df2.index >= '2020-07-01']

# Set the figure size
fig, ax = plt.subplots(figsize=(15, 7))

# Plot the ratio
ax.plot(df1_filtered.index, df1_filtered['Ratio'], color='b', label='Oil/Gas Ratio')

# Set the title and axis labels and sizes
ax.set_title('Oil/Gas Futures Ratio', fontsize=16)
ax.set_xlabel('Year-Month', fontsize=15)
ax.set_ylabel('Ratio', fontsize=15)
ax.tick_params(axis='both', labelsize=15)
ax.legend(loc='upper left', fontsize=15)

# Show the plot
plt.show()
```
